```python
import pandas as pd
```

```python
df_17 = pd.read_csv('2017-tw-foreign-nationality-reshape.csv')
df_22 = pd.read_csv('2022-tw-foreign-nationality-reshape.csv')
```

## Countries with most immigrants in Taiwan as a whole

```python
# total, top 10 places
df_17.groupby('Country_EN').sum().sort_values('總 計 Total', ascending=False).head(10)
```

```python
df_22.groupby('Country_EN').sum().sort_values('總 計 Total', ascending=False).head(10)
```

```python
# Calculating the percent of the total number of foreigners in Taiwan that immigrated from Vietnam, the Philippines, Indonesia, Thailand or Malaysia
(225949 + 215491 + 143351 + 65947 + 23565) / sum(df_22['總 計 Total'])
```

## Countries with most immigrants in Taoyuan
